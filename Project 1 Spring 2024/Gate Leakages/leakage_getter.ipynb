# NAND

```python
import subprocess
import os

netlist_template = """
*Leakage current/power in NAND2 cell
.INCLUDE ../model/32nm_HP.pm


.PARAM Lmin=22n
.PARAM Wmin=22n
.PARAM XX=1
.PARAM tr=50p

Mp1	nodeZ	nodea	node1	node1	pmos	W={{2*XX*Wmin}}	L={{Lmin}}
Mp2	nodeZ	nodeb	node1	node1	pmos	W={{2*XX*Wmin}}	L={{Lmin}}

Mn1	node4	nodeb	0	0	nmos	W={{2*XX*Wmin}}	L={{Lmin}}
Mn2	nodeZ	nodea	node4	0	nmos	W={{2*XX*Wmin}}	L={{Lmin}}

Vdd 	node1 	0 	1

Vina nodea 0 0
Vinb nodeb 0 0

*Definition of type of analysis
.CONTROL
foreach Vddbasic 1.0
*let Vddbasic = 1
echo _|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_
echo _|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_
echo Vdd basic = $Vddbasic
alter Vdd = $Vddbasic
foreach width 22n
    echo
	echo _|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_|_
	echo
	echo WMIN=$width
	echo

	    foreach input 0 $Vddbasic
 	    	alter Vina = $input
 	    		foreach input 0 $Vddbasic
				alter Vinb = $input
			                    echo DC Analysis at 85C and 145C
		                        	dc TEMP 85 86 60
			                    		print V(nodeZ) V(nodea) V(nodeb) I(Vdd) I(Vina) I(Vinb)




	        	end
	    end
end
end
.ENDC
.END

"""


netlist_content = netlist_template.format()

filename = f"./nand_print.net"

with open(filename, 'w') as file:
    file.write(netlist_content)

p = subprocess.run(["ngspice", filename] , encoding='utf-8', 
stdout=subprocess.PIPE)
windstr = p.stdout
out = windstr.split('\n')
```

```python
# make a csv
with open('nand_print.csv', 'w') as file:
    file.write('V(nodeZ),V(nodea),V(nodeb),I(Vdd),I(Vina),I(Vinb)\n')
    for i in range(14, len(out), 15):
        temp = out[i:i+15]
        temp = [x.split('=')[-1] for x in temp[9:15]]
        file.write(','.join(temp) + '\n')
```
